## Reading an image file into a DataFrame

```python
from pyspark.ml import image
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("img_process").getOrCreate()
```

```python
import os

image_files_dir = os.path.join(os.getcwd(),"datasets/cctvVideos/train_images/")
```

```python
img_df = spark.read.format("image").load(image_files_dir)
```

```python
img_df.printSchema()
```

```python
img_df.select("image.origin","image.height","image.width","image.nChannels","image.mode","label").show()
```
